```python
stop_words = stopwords.words('english')
stop_words=set(stop_words)
print(stop_words)
```

```python
def clean_tokens(text):
    # 1 create tokens
    tokens = word_tokenize(text)
    # 2 lower case
    tokens = [w.lower() for w in tokens]
    # 3 remove punctuations
    stripped = [word for word in tokens if word.isalpha()]
    # 4 remove stop_words
    stop_words = set(stopwords.words('english'))
    words = [w for w in stripped if not w in stop_words]
    
    clean_text = ' '.join(words)
    
    return clean_text
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    tokens = word_tokenize(text) 
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)


df['lemmatized_remarks'] = df['processed_remarks'].apply(lemmatize_text)
```

```python
clf=LogisticRegression()
```

```python
%%time
clf.fit(X_train,y_train)
```

```python
%%time
print("Precision Score for Logistic Regression: %s" % precision_score(y_test,clf.predict(X_test),average='micro'))
print("Recall Score for Logistic Regression: %s" % recall_score(y_test,clf.predict(X_test),average='micro'))
print("AUC Score for Logistic Regression: %s" % roc_auc_score(y_test,clf.predict_proba(X_test),multi_class='ovr',average='macro'))
f1_score =f1_score(y_test,clf.predict(X_test),average="weighted")
print("F1 Score for Logistic Regression: %s" % f1_score)
print("Accuracy Score for Logistic Regression: %s" % accuracy_score(y_test,clf.predict(X_test)))
```

```python
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
from sklearn.model_selection import cross_val_predict
y_pred_cv = cross_val_predict(clf, X, y, cv=5)
cm_cv = confusion_matrix(y, y_pred_cv)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_cv, annot=True, fmt='d', cmap='BuPu', xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
